```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
y1, sr1 = librosa.core.load('jeremaudios/E_finger.wav')
y2, sr2 = librosa.core.load('jeremaudios/E_pick.wav')
y1 = y1[:3*sr1]
y2 = y2[:3*sr2]
n = 13
mfcc1 = librosa.feature.mfcc(y=y1, sr=sr1,n_mfcc =n,  dct_type=1)
mfcc2 = librosa.feature.mfcc(y=y2, sr=sr2,n_mfcc =n, dct_type=1)
mfcc_diff = mfcc1-mfcc2

```

```python
# Вывод графика MFCC для первого файла
plt.figure(figsize=(10, 4))
y_coords = np.linspace(0, n, num=mfcc1.shape[0])
librosa.display.specshow(mfcc1, x_axis='time',y_axis='linear', y_coords=y_coords)
plt.colorbar()
plt.title('MFCC E_Finger')
plt.ylabel('Coefficients')
plt.tight_layout()
plt.show()
```

```python
# Вывод графика MFCC для первого файла
plt.figure(figsize=(10, 4))
y_coords = np.linspace(0, n, num=mfcc1.shape[0])
librosa.display.specshow(mfcc2, x_axis='time',y_axis='linear', y_coords=y_coords)
plt.colorbar()
plt.title('MFCC E_Pick')
plt.ylabel('Coefficients')
plt.tight_layout()
plt.show()
```

```python
# Вывод графика MFCC для первого файла
plt.figure(figsize=(10, 4))
y_coords = np.linspace(0, n, num=mfcc1.shape[0])
librosa.display.specshow(mfcc_diff, x_axis='time',y_axis='linear', y_coords=y_coords)
plt.colorbar()
plt.title('MFCC E_Diff')
plt.ylabel('Coefficients')
plt.tight_layout()
plt.show()
```

```python
# Feature Generation
# MFCCs
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
mfccs1_mean=np.mean(mfcc1,axis=1)
mfccs1_std=np.std(mfcc1,axis=1)
mfccs2_mean=np.mean(mfcc2,axis=1)
mfccs2_std=np.std(mfcc2,axis=1)
mfccs_diff_mean = np.mean(mfcc_diff,axis=1)
mfccs_diff_std = np.std(mfcc_diff,axis=1)
mfccs_diff_mean2 = mfccs1_mean - mfccs2_mean
mfccs_diff_std2 = mfccs1_std - mfccs2_std
coeffs=np.arange(n)
```

```python
#MEAN
plt.figure(figsize=(15,5))
plt.title('Mean MFCCs Finger')
plt.ylabel('Value')
plt.xlabel('Coefficients')
ax = sns.barplot(x=coeffs,y=mfccs1_mean)
for i, v in enumerate(mfccs1_mean):
    ax.text(i,v, f'{v:.0f}', color='black', ha='center')
plt.show()

plt.figure(figsize=(15,5))
plt.title('Mean MFCCs Pick')
ax = sns.barplot(x=coeffs,y=mfccs2_mean)
for i, v in enumerate(mfccs2_mean):
    ax.text(i,v, f'{v:.0f}', color='black', ha='center')
plt.show()
```

```python
#SD
plt.figure(figsize=(15,5))
plt.title('SD MFCCs Finger')
plt.ylabel('Value')
plt.xlabel('Coefficients')
ax = sns.barplot(x=coeffs,y=mfccs1_std)
for i, v in enumerate(mfccs1_std):
    ax.text(i,v, f'{v:.0f}', color='black', ha='center')
plt.show()


plt.figure(figsize=(15,5))
plt.title('SD MFCCs Pick')
plt.ylabel('Value')
plt.xlabel('Coefficients')
ax = sns.barplot(x=coeffs,y=mfccs2_std)
for i, v in enumerate(mfccs2_std):
    ax.text(i,v, f'{v:.0f}', color='black', ha='center')
plt.show()
```

```python
#Diff
plt.figure(figsize=(15,5))
plt.title('Mean MFCCs Diff')
plt.ylabel('Value')
plt.xlabel('Coefficients')
ax = sns.barplot(x=coeffs,y=mfccs_diff_mean2)
for i, v in enumerate(mfccs_diff_mean2):
    ax.text(i,v, f'{v:.0f}', color='black', ha='center')
plt.show()

plt.figure(figsize=(15,5))
plt.title('SD MFCCs Diff')
plt.ylabel('Value')
plt.xlabel('Coefficients')
ax = sns.barplot(x=coeffs,y=mfccs_diff_std2)
for i, v in enumerate(mfccs_diff_std2):
    ax.text(i,v, f'{v:.0f}', color='black', ha='center')
plt.show()
```

```python
# A-D
import os
n = 10
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 6))
fig.subplots_adjust(hspace=0.5)
folder_path = 'jeremaudios'
file_list = os.listdir(folder_path)
coeffs=np.arange(n)
axes = axes.ravel()

for i in range(6):
    current_file = file_list[i]
    y1, sr1 = librosa.load("jeremaudios/"+current_file)
    y1 = y1[:3*sr1]
    mfcc_final = librosa.feature.mfcc(y=y1, sr=sr1, n_mfcc=n, dct_type=3)
    mfccs_final_mean = np.mean(mfcc_final, axis=1)
    sns.barplot(ax=axes[i],x=coeffs,y=mfccs_final_mean)
    axes[i].set_title(f'Mean MFCCs {current_file}')
    axes[i].set_xticks(np.arange(0, n, 2))
fig.show()
```

```python
# EE - G
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(12, 8))
fig.subplots_adjust(hspace=0.5)
folder_path = 'jeremaudios'
file_list = os.listdir(folder_path)
coeffs=np.arange(n)
axes = axes.ravel()

for i in range(6,14):
    current_file = file_list[i]
    y1, sr1 = librosa.load("jeremaudios/"+current_file)
    y1 = y1[:3*sr1]
    mfcc_final = librosa.feature.mfcc(y=y1, sr=sr1, n_mfcc=n, dct_type=3)
    mfccs_final_mean = np.mean(mfcc_final, axis=1)
    sns.barplot(ax=axes[i-6],x=coeffs,y=mfccs_final_mean)
    axes[i-6].set_title(f'Mean MFCCs {current_file}')
    axes[i-6].set_xticks(np.arange(0, n, 2))
fig.show()
```

```python
## Создание датафреймов для всех медиан и отклонений.
import pandas as pd
n = 39
folder_path = 'jeremaudios'
file_list = os.listdir(folder_path)

df_mean_all = pd.DataFrame(index=range(n), columns=range(len(file_list)))
df_std_all = pd.DataFrame(index=range(n), columns=range(len(file_list)))
for i in range(len(file_list)):
    y1, sr1 = librosa.load("jeremaudios/" + file_list[i])
    y1 = y1[:3*sr1]
    mfccs_df = librosa.feature.mfcc(y=y1, sr=sr1, n_mfcc=n, dct_type=3)
    mfccs_df_mean = np.mean(mfccs_df, axis=1)
    mfccs_df_std = np.std(mfccs_df, axis=1)
    # Generate the chroma Dataframe
    for k in range(n):
        df_mean_all.at[k, i] = mfccs_df_mean[k]
    for k in range(n):
        df_std_all.at[k, i] = mfccs_df_std[k]

df_mean_all.columns = file_list
df_std_all.columns = file_list
```

```python
means_x2 = []
means = df_mean_all.sum().to_numpy()
for i in range(0,len(means)-1,2):
    means_x2.append([means[i],means[i+1]])
```

```python
stds_x2 = []
stds = df_std_all.sum().to_numpy()
for i in range(0,len(stds)-1,2):
    stds_x2.append([stds[i],stds[i+1]])
```

```python
for i in range(len(means_x2)):
    plt.scatter(i,means_x2[i][0],c='red')
    plt.scatter(i,means_x2[i][1],c='black')
plt.title('Mean value for finger and pick')
plt.ylabel('Value')
plt.xlabel('Pitches')
plt.legend(['finger', 'pick'])
plt.show()
```

```python
for i in range(len(stds_x2)):
    plt.scatter(i,stds_x2[i][0],c='red')
    plt.scatter(i,stds_x2[i][1],c='black')
plt.title('Sd value for finger and pick')
plt.ylabel('Value')
plt.xlabel('Pitches')
plt.legend(['finger', 'pick'])
plt.show()
```
